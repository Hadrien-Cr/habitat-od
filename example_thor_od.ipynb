```python
from pathlib import Path
from thor_od.data_gen import load_dataset
from common.utils.plot_utils import plot_segmentation, plot_object_detection

loaded_dataset, classes, colors = load_dataset(Path("data/procthor"))
```

```python
img, labels = loaded_dataset["train"][1]

_img = plot_segmentation(
    img=img, 
    prediction=[], 
    labels=labels, 
    classes=classes,
    colors=colors
)

import matplotlib.pyplot as plt

plt.imshow(_img)
plt.axis('off')
```

```python
from ultralytics import SAM, YOLOWorld

detection_model = YOLOWorld("yolov8x-world.pt")
detection_model.set_classes(classes=list(classes.values()))

segmentation_model = SAM("sam_b.pt")
```

```python
det = detection_model.predict(img)[0].boxes
xyxy = det.xyxy
conf = det.conf
cls = det.cls

pred = []

for i in range(len(cls)):
    pred.append((int(cls[i]), (int(xyxy[i][0]), int(xyxy[i][1]), int(xyxy[i][2]), int(xyxy[i][3])), conf[i].item()))


_img = plot_object_detection(
    img=img, 
    prediction=pred, 
    labels=[], 
    classes=classes,
    colors=colors
)

import matplotlib.pyplot as plt

plt.imshow(_img)
plt.axis('off')
```

```python
seg = segmentation_model.predict(img)[0].masks
binary_masks = seg.data
cls = list(range(len(binary_masks)))

pred = []

for i in range(len(cls)):
    pred.append((0, binary_masks[i].cpu().numpy(), 0.0))

_img = plot_segmentation(
    img=img, 
    prediction=pred, 
    labels=[], 
    classes=classes,
    colors=colors
)

import matplotlib.pyplot as plt

plt.imshow(_img)
plt.axis('off')
```
